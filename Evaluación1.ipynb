```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se lee 'vid18_180219.dat' y se le dice que esta 
#separado por comas "delimiter = ',' "
d0 = pd.read_csv('vid18_180219.dat', delimiter = ',')
df1 = pd.DataFrame(d0)
df1
```

```python
#Transformando a el formato de fecha
df1["TIMESTAMP"] = pd.to_datetime(df1.TIMESTAMP)
```

```python
#Para limpiar el dataframe anterior (df2) y solo dejar
#la columna "TIMESTAMP" y "AirTC_Avg"
df2 = df1.filter(["TIMESTAMP", "AirTC_Avg"])
df2.tail()
```

```python
#Aprovechando de que la columna de la fecha tiene dicho
#formato, se crean columnas con el ao, mes, da y hora
#para trabajar ms fcil con la tabla de datos
df2["Month"] = df2["TIMESTAMP"].dt.month
df2["Year"] = df2["TIMESTAMP"].dt.year
df2["Hour"] = df2["TIMESTAMP"].dt.hour
df2["Day"] = df2["TIMESTAMP"].dt.day
df2.head(), df2.tail()
```

```python
#Se calcula el promedio diario de la temperatura, para
#as despus calcular los UF24 de manera ms fcil
#   Se empieza del dato 24937 porque es el dato
#donde se encuentra el 1ro de Noviembre

Hora = 0
Count = 0
Temp0 = 0
fecha = []
prom = []
for i in range(24937, len(df2)):
    if(df2["Hour"][i] == Hora):
        Count = Count + 1
        Temp0 = Temp0 + df2["AirTC_Avg"][i]
        Hora = df2["Hour"][i]
    else: 
        prom.append(Temp0/Count)
        fecha.append(df2["TIMESTAMP"][i])
        Count = 1
        Hora = df2["Hour"][i]
        Temp0 = df2["AirTC_Avg"][i]
prom.append(Temp0/Count)
fecha.append(df2["TIMESTAMP"][i])
```

```python
#Se crea un nuevo dataframe donde tengo el promedio 
#diario ya con fecha diaria
df3 = pd.DataFrame(
    {"Fecha" : fecha,
     "Temperatura Promedio" : prom},
      )
```

```python
#Como se hizo anteriormente, se le mete el mes, ao y 
#da al datafram (df3)
df3["Month"] = df3["Fecha"].dt.month
df3["Year"] = df3["Fecha"].dt.year
df3["Day"] = df3["Fecha"].dt.day
df3.head(), df3.tail()
```

```python
#Se calculan los UF24 sin ser acumulados
day = 1
uf = 0
UF = []
date = []
for i in range(len(df3)):
    if(df3["Day"][i] == day):
        day = df3["Day"][i]
        if (df3["Temperatura Promedio"][i] < 1.4):
            continue
        elif(1.5 < df3["Temperatura Promedio"][i] < 2.4):
            uf = uf + 0.5
        elif(2.5 < df3["Temperatura Promedio"][i] < 9.1):
            uf = uf + 1
        elif(9.2 < df3["Temperatura Promedio"][i] < 12.4):
            uf = uf + 0.5
        elif(12.5 < df3["Temperatura Promedio"][i] < 15.9):
            continue
        elif(16 < df3["Temperatura Promedio"][i] < 18):
            uf = uf - 0.5
        elif(df3["Temperatura Promedio"][i] > 18):
            uf = uf - 1
    else:
        UF.append(uf)
        date.append(df3["Fecha"][i])
        uf = 0
        day = df3["Day"][i]
UF.append(uf)
date.append(df3["Fecha"][i])        
```

```python
#Se crea un DataFrame con los datos de los UF24 para su
#posterior anlisis
DatTable = pd.DataFrame(
    {"Date" : date,
     "Unit of cold" : UF},
)
```

```python
#Calculando las temperaturas mximas diaras
day = 1
temp0 = 0
tempMax = []
#date = []
for i in range(len(df3)):
    if(df3["Day"][i] == day):
        day = df3["Day"][i]
        if(df3["Temperatura Promedio"][i] > temp0):
            temp0 = df3["Temperatura Promedio"][i]
    else:
        tempMax.append(temp0)
        #date.append(df3["Fecha"][i])
        temp0 = 0
        day = df3["Day"][i]
tempMax.append(temp0)
#date.append(df3["Fecha"][i])        
```

```python
#Calculando las temperaturas mnimas diaras
day = 1
temp0 = 100000
tempMin = []
#date = []
for i in range(len(df3)):
    if(df3["Day"][i] == day):
        day = df3["Day"][i]
        if(df3["Temperatura Promedio"][i] < temp0):
            temp0 = df3["Temperatura Promedio"][i]
    else:
        tempMin.append(temp0)
        #date.append(df3["Fecha"][i])
        temp0 = 10000
        day = df3["Day"][i]
tempMin.append(temp0)
#date.append(df3["Fecha"][i])  
```

```python
#Aadiendo la temperatura mxima y mnima a mi DataFrame
#con los datos de UF
DatTable["Tmax"] = tempMax
DatTable["Tmin"] = tempMin
```
